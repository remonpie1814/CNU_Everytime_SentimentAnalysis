```python
donga_senti= pd.read_csv("/Users/KBS/Documents/CNU_Everytime_SentimentAnalysis/donga_senti.csv",sep="\t",encoding="utf-8")

tokenizer = Okt()
# 
#stop_words = ['','','','','','','','','','','','','','','','','','','','','','','','']
#donga_senti['tokenized'] = donga_senti['tokenized'].apply(lambda x: [item for item in x if item not in stop_words and len(item) > 1])

donga_senti['tokenized'] = donga_senti['title'].apply(tokenizer.nouns)
donga_senti['tokenized'] = donga_senti['tokenized'].apply(lambda x: [item for item in x if item not in stop_words and len(item) > 1])

donga_Politics = donga_senti[donga_senti['section'].str.contains('Politics')] ####
donga_Politics[donga_Politics['section']=='title']['tokenized'].values
donga_Politics = np.hstack(donga_Politics[donga_Politics['section']=='Politics']['tokenized'].values) ####
donga_Politics = Counter(donga_Politics)#
donga_Politics = donga_Politics.most_common(100)#
donga_Politics = dict(donga_Politics)#
wc = WordCloud(
    font_path= '/Library/Fonts/NotoSansKR-Bold.otf',
    background_color='white',
    width=1600,
    height=1600,
)
wc_img = wc.generate_from_frequencies(donga_Politics) #
wc_img.to_file('donga_wordcloud_Politics.jpg')
plt.imshow(wc, interpolation = 'bilinear')
```
