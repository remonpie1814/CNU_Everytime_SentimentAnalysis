```python
for idx, t in enumerate(data['etime']):
    if "분" in t:
        temp = t.split("분")
        d = datetime.datetime.now()-datetime.timedelta(seconds=int(temp[0])*60)
        data["etime"][idx] = d.strftime('%Y/%m/%d %H:%M')
    elif "방금" in t:
        d = datetime.datetime.now()
        data["etime"][idx]=d.strftime('%Y/%m/%d %H:%M') 
    elif len(t)<12:
        data["etime"][idx] = "2022/"+data["etime"][idx]
        
```
